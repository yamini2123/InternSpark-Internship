```python
df.drop_duplicates(inplace=True)
df["Purchase Date"] = pd.to_datetime(df["Purchase Date"])
df.isnull().sum()
```

```python

df.head()
```

```python
# Extract Year
df["Year"] = df["Purchase Date"].dt.year

# Extract Month
df["Month"] = df["Purchase Date"].dt.month_name()

# Extract Day
df["Day"] = df["Purchase Date"].dt.day_name()

# Display first 5 rows
df.head()
```

```python
df["Average Purchase"] = df["Total Purchase Amount"] / df["Quantity"]

df.head()
```

```python
print(df[["Purchase Date", "Year", "Month", "Day", "Average Purchase"]].head())
```

```python
snapshot_date = df["Purchase Date"].max() + pd.Timedelta(days=1)
rfm = df.groupby("Customer ID").agg({
    "Purchase Date": lambda x: (snapshot_date - x.max()).days,
    "Customer ID": "count",
    "Total Purchase Amount": "sum"
})

rfm.columns = ["Recency", "Frequency", "Monetary"]

rfm.head()
```

```python
rfm["R"] = pd.qcut(rfm["Recency"], 4, labels=[4,3,2,1])
rfm["F"] = pd.qcut(rfm["Frequency"].rank(method="first"), 4, labels=[1,2,3,4])
rfm["M"] = pd.qcut(rfm["Monetary"], 4, labels=[1,2,3,4])

rfm.head()
```

```python
#RFM SCORE 
rfm["RFM Score"] = (
    rfm["R"].astype(str) +
    rfm["F"].astype(str) +
    rfm["M"].astype(str)
)

rfm.head()
```

```python
# Create Customer Segments
def segment_customer(row):
    if row["R"] == 4 and row["F"] == 4 and row["M"] == 4:
        return "Champions"
    elif row["R"] >= 3 and row["F"] >= 3:
        return "Loyal Customers"
    elif row["R"] >= 3:
        return "Potential Loyalists"
    elif row["R"] <= 2 and row["F"] >= 3:
        return "At Risk"
    else:
        return "Lost Customers"

rfm["Segment"] = rfm.apply(segment_customer, axis=1)

rfm.head()
```

# Customer Segment Profiles

- **Champions:** Recently purchased, purchase frequently, and spend the most. These are the most valuable customers.
- **Loyal Customers:** Regular buyers with consistent purchase behavior. They have strong long-term potential.
- **Potential Loyalists:** Recent customers who can become loyal with proper engagement.
- **At Risk:** Customers who purchased in the past but have become less active and may churn.
- **Lost Customers:** Customers who have not purchased for a long time and require re-engagement strategies.
